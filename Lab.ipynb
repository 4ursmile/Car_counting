```python
import cv2
import numpy as np
import os
import time
from ultralytics import YOLO
class ObjectDetection:
    def __init__(self, model_path="./YOLOV8N-Vehicle/train/weights/best.pt", conf = 0.2, iou = 0.7, device = None):
        print("Dectector model is loading...")
        self.model = YOLO(model_path)
        self.conf = conf
        self.iou = iou
        self.device = device
        print("Dectector model is ready now")
    def predict(self, frame):
        results = self.model.predict(frame, conf=self.conf, iou=self.iou, device=self.device)
        results[0] = results[0].numpy()
        return results[0].boxes.xyxy
```

```python
width = 848
height = 477
dim = (width, height)
def image_resize(img, dim):
    return cv2.resize(img, dim, interpolation=cv2.INTER_AREA)

def image_preprocess(img):
    grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(grey, (5, 5), 0)
    equal = cv2.equalizeHist(blur)
    return equal
```

```python
video_capture = cv2.VideoCapture("./Videos/camera2.mp4")
```

```python
fps = video_capture.get(cv2.CAP_PROP_FPS)
real_time = round(1000/fps)
target_fps = 20

counter = 0
targer_counter = 1;
detector = ObjectDetection()
cap_x, cap_y, cap_w, cap_h = 400, 150, 420, 315 
show_cap = True
```

```python
while True:
    if counter == targer_counter:
        counter = 0;
        ret = video_capture.grab()
        continue
    counter += 1
    ret, frame = video_capture.read()
    if ret:
        start_time = time.time();
        frame = image_resize(frame, dim)
        roi = frame[cap_y:cap_y + cap_h, cap_x:cap_x + cap_w]
        if show_cap:
            cv2.rectangle(frame, (cap_x, cap_y), (cap_x + cap_w, cap_y + cap_h), (255, 255, 0), 2)
        results = detector.predict(roi)
        for xyxy in results:
            ux, uy, lx, ly = xyxy.astype(int)
            cv2.rectangle(roi, (ux, uy), (lx, ly), (0, 255, 0), 2)
        #cv2.rectangle(an_frame, (cap_x, cap_y), (cap_x + cap_w, cap_y + cap_h), (0, 255, 0), 2)
        end_time = time.time()
        # eslap_time = (end_time - start_time)*1000
        # if real_time > eslap_time:
        #     time.sleep((real_time - eslap_time)/1000)
        # else:
        #     for i in range(frame_skip-1):
        #         _, _ = video_capture.read()
        cv2.imshow("frame", frame)
    else:
        video_capture.set(cv2.CAP_PROP_POS_FRAMES, 0)
        continue
    waiting_time = int(1000/fps)
    key = cv2.waitKey(1)
    if key == ord('x'):
        break

video_capture.release()
cv2.destroyAllWindows()
```
